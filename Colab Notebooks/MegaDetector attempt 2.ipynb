```python
import os
import shutil
import math
from multiprocessing import Pool

def copy_files_with_structure(file_paths, dest_dir):
    #print(dest_dir)
    for source_file_path in file_paths:
        relative_path = source_file_path.replace("/content/drive/MyDrive/Project", "")
        dest_file_path = dest_dir + relative_path
        print(dest_file_path)
        dest_folder_path = os.path.dirname(dest_file_path)
        os.makedirs(dest_folder_path, exist_ok=True)
        shutil.copy(source_file_path, dest_file_path)

# Define batch folder base directory
batch_folder_base = '/content/drive/MyDrive/Project/Batch Folders'

# Define the batch size (number of files per batch)
batch_size = 20100

# Calculate the number of batches
num_batches = math.ceil(len(species_only_paths_list) / batch_size)

# Create batch folders
for i in range(1, num_batches):
    batch_folder = os.path.join(batch_folder_base, f'batch_{i}')
    os.makedirs(batch_folder, exist_ok=True)

# Parallelize file copying
with Pool() as pool:
    for i in range(1, num_batches):
        start_index = i * batch_size
        end_index = min((i + 1) * batch_size, len(species_only_paths_list))
        batch_files = species_only_paths_list[start_index:end_index]
        dest_batch_folder = os.path.join(batch_folder_base, f'batch_{i}')
        pool.apply_async(copy_files_with_structure, args=(batch_files, dest_batch_folder))

    pool.close()
    pool.join()  # Wait for all processes to finish
```

```python
def copy_files_with_structure(file_paths, dest_dir):
    num_files = len(file_paths)
    for i, source_file_path in enumerate(file_paths):
        relative_path = source_file_path.replace("/content/drive/MyDrive/Project", "")
        dest_file_path = dest_dir + relative_path
        print(dest_file_path)
        dest_folder_path = os.path.dirname(dest_file_path)
        os.makedirs(dest_folder_path, exist_ok=True)
        shutil.copy(source_file_path, dest_file_path)

        # Print progress as a percentage
        progress = ((i + 1) / num_files) * 100
        print(f"Batch progress: {progress:.2f}% ({i + 1}/{num_files})")

# Define batch folder base directory
batch_folder_base = '/content/drive/MyDrive/Project/Batch Folders'

# Define the batch size (number of files per batch)
batch_size = 20100

# Calculate the number of batches
num_batches = math.ceil(len(species_only_paths_list) / batch_size)

# Create batch folders
for i in range(1, num_batches):
    batch_folder = os.path.join(batch_folder_base, f'batch_{i}')
    os.makedirs(batch_folder, exist_ok=True)

# Parallelize file copying
with Pool() as pool:
    for i in range(1, num_batches):
        start_index = i * batch_size
        end_index = min((i + 1) * batch_size, len(species_only_paths_list))
        batch_files = species_only_paths_list[start_index:end_index]
        dest_batch_folder = os.path.join(batch_folder_base, f'batch_{i}')
        pool.apply_async(copy_files_with_structure, args=(batch_files, dest_batch_folder))

    pool.close()
    pool.join()  # Wait for all processes to finish
```
