Have 96853 crops in total according to the length of these dfs

Now, have 108,037 crops with included rarer species from leftrover batches 2&3

**Combining paths dfs into one big df**

```python
combined_df_paths = pd.concat([sorted_leftover_batch1_df_paths,
                               MSV_crops_df_paths,
                               MSV_crops_df_paths2,
                               chunk0_crops_df_paths,
                               chunk0_crops_df_paths2,
                               chunk0_crops_df_paths3,
                               leftover_include_df_paths], ignore_index=True)
```

```python
print(len(combined_df_paths))
combined_df_paths.head()
```

**Mapping crop paths to original image paths so that they can be matched with corresponding species label**

```python
#original paths included in this table

data_for_MD_df = pd.read_csv("/content/drive/MyDrive/Project/data_for_MD.csv")

print(len(data_for_MD_df))
data_for_MD_df.head()
```

```python
print(chunk0_crops_df_paths['full_img_dir'][0])
```

```python
print(data_for_MD_df['SourceFile'][0])
```

Changing species_true_only_df3 paths to drive paths

```python
data_for_MD_df['SourceFile'] = data_for_MD_df['SourceFile'].str.replace('D:/', '/content/drive/MyDrive/Project/')
data_for_MD_df.head()
```

**Merging dataframes together**

```python
merged_df = data_for_MD_df.merge(
    combined_df_paths[['crop_path', 'md_conf', 'full_img_dir']].rename(columns={'full_img_dir': 'SourceFile'}),
    how='left', on='SourceFile'
)


merged_df.head()
```

```python
NaN_counter = 0

for index, row in merged_df.iterrows():
    if pd.isna(row['crop_path']) or pd.isna(row['md_conf']):
        NaN_counter += 1

print(NaN_counter)
```

```python
print(len(merged_df))
```

Dropping NaN values from the merged dataframe

```python
filtered_df = merged_df.dropna(subset=['crop_path', 'md_conf'])
print(len(filtered_df))
filtered_df.head()
```
